### Write a python program to Implement shopping cart using linked-list. The user should be able to add item, remove item, display all the items, calculate total amount of the cart and count total number of items in a cart.Each Item details contains (Item id , name, quantity, price).

```python
class Node:
  def __init__(self,id = 0, name=None, quantity = 0,price= 0):
    self.id = id
    self.name = name
    self.quantity = quantity
    self.price = price
    self.next = None

class LinkedList:
  def __init__ (self):
    self.start = None

  def atEnd(self, id, name, quantity,price):
    p = Node(id, name,quantity,price)
    if self.start is None:
      self.start = p
    else:
      q = self.start
      while q.next is not None:
        q = q.next
      q.next = p
    print("Product added !")

  def listPrint(self):
    q = self.start
    print('\t\t\tSHOPPING CART')
    print("ID\tName\tQuantity\tPrice")
    print("-" * 40)
    while q is not None:
      print(f"{q.id}\t{q.name}\t{q.quantity}\t\t${q.price}")
      q = q.next

  def removeNode(self, id):
    if self.start is None:
      print("Cart is empty!")
      return
    
    if self.start.id == id:
      self.start = self.start.next
      print("Product removed!")
      return
    
    q = self.start
    while q.next is not None:
      if q.next.id == id:
        q.next = q.next.next
        print("Product removed!")
        return
      q = q.next
    print("Product not found!")

  def sumofcart(self):
    q = self.start
    if q is None:
      return 0
    sum = 0
    while q is not None:
      sum += q.price * q.quantity
      q = q.next
    return sum

  def countItems(self):
    count = 0
    q = self.start
    while q is not None:
      count += q.quantity
      q = q.next
    return count
```

```python
import sys

l1 = LinkedList()
while True:
  print("\n1. Add an item to the cart")
  print("2. Delete an item from the cart")
  print("3. Display Items in the cart")
  print("4. Total amount")
  print("5. Exit")

  try:
      choice = int(input("\nEnter the choice: "))
      if(choice == 1):
          try:
              id = int(input("Enter Product id: "))
              name = input("Enter product name: ")
              price = float(input("Enter product price per item(in Rs): "))
              quantity = int(input("Enter quantity(in units): "))
              l1.atEnd(id, name, quantity, price)
          except ValueError:
              print("Please enter valid numeric values")
      elif choice == 2:
          try:
              id = int(input("Enter the product id you want to remove: "))
              if(l1.start is None):
                  print("Cart is empty")
              else:
                  output = l1.removeNode(id)
                  if output:
                      print("Product deleted!")
                  else:
                      print("Product not found")
          except ValueError:
              print("Please enter a valid product id")
      elif choice == 3:
          if(l1.start is None):
              print("Cart is empty")
          else:
              l1.listPrint()
      elif choice == 4:
          if(l1.start is None):
              print("Cart is empty")
          else:
              sum = float(l1.sumofcart())
              print("\nTotal Amount: ", sum)
      elif choice == 5:
          print("Thank you for shopping!")
          sys.exit()
      else:
          print("Please enter a number between 1 and 5")
  except ValueError:
      print("Please enter a valid number")
```

###Infix to postfix conversion
Create an empty stack called opstack for keeping operators. Create an empty list for output.

1. Convert the input infix string to a list by using the string method split.

2. Scan the token list from left to right.

3. If the token is an operand, append it to the end of the output list.

4. If the token is a left parenthesis, push it on the opstack.

5. If the token is a right parenthesis, pop the opstack until the corresponding left parenthesis is removed. Append each operator to the end of the output list.

6. If the token is an operator, *, /, +, or -, push it on the opstack. However, first remove any operators already on the opstack that have higher or equal precedence and append them to the output list.

7. When the input expression has been completely processed, check the opstack. Any operators still on the stack can be removed and appended to the end of the output list.



```python
class Stack:
  pass
def infixToPostfix(infixexpr):
    prec = {}
    prec["*"] = 3
    prec["/"] = 3
    prec["+"] = 2
    prec["-"] = 2
    prec["("] = 1

    opStack = Stack()
    postfixList = []

    tokenList = infixexpr.split()

    for token in tokenList:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "0123456789":
          postfixList.append(token)
        elif token == '(':
          opStack.push(token)
        elif token == ')':
            topToken = opStack.pop()
            while topToken != '(':
                postfixList.append(topToken)
                topToken = opStack.pop()
        else:
            while (not opStack.isEmpty()) and (prec[opStack.peek()] >= prec[token]):
                  postfixList.append(opStack.pop())
            opStack.push(token)

    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    return " ".join(postfixList)
```

```python
print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
```

###Doubly linked list
1. insert into odered linked list
2. display linked list using recursive method
3. count number of nodes using recursive method

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def insert_ordered(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            return
        
        if data < self.head.data:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            return
        
        current = self.head
        while current.next and current.next.data < data:
            current = current.next
            
        new_node.next = current.next
        new_node.prev = current
        
        if current.next:
            current.next.prev = new_node
        current.next = new_node
    
    def display_recursive(self, node=None):
        if node is None:
            node = self.head
            
        if node:
            print(node.data, end=" ")
            self.display_recursive(node.next)
        elif node == self.head:
            print("List is empty")
    
    def count_nodes_recursive(self, node=None):
        if node is None:
            node = self.head
            
        if node is None:
            return 0
        
        return 1 + self.count_nodes_recursive(node.next)

# Example usage
dll = DoublyLinkedList()
dll.insert_ordered(5)
dll.insert_ordered(3)
dll.insert_ordered(8)
dll.insert_ordered(1)
dll.insert_ordered(7)

print("Linked List:")
dll.display_recursive()
print("\nNumber of nodes:", dll.count_nodes_recursive())
```

## Tree implementation using Object Oriented Programming in **Python**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None
        
    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursive(self.root, data)
            
    def _insert_recursive(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursive(node.left, data)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursive(node.right, data)
                
    def inorder_traversal(self, node):
        if node:
            self.inorder_traversal(node.left)
            print(node.data, end=' ')
            self.inorder_traversal(node.right)
            
    def preorder_traversal(self, node):
        if node:
            print(node.data, end=' ')
            self.preorder_traversal(node.left)
            self.preorder_traversal(node.right)
            
    def postorder_traversal(self, node):
        if node:
            self.postorder_traversal(node.left)
            self.postorder_traversal(node.right)
            print(node.data, end=' ')
```
